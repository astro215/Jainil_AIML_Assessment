```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from keras.preprocessing.sequence import TimeseriesGenerator
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np

# Load data
data = pd.read_csv("train_df_quantify.csv")  # Replace "your_data.csv" with your file path



# Label encoding for categorical variables
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

# Normalize numerical variables
scaler = MinMaxScaler()
data[["Sourcing Cost"]] = scaler.fit_transform(data[["Sourcing Cost"]])

# Filter data for training (July 2020 to May 2021)
train_data = data[(data['Year'] == 2020) & (data['Month'] >= 7) | (data['Year'] == 2021) & (data['Month'] <= 5)]

# Prepare sequences for LSTM
X_train = train_data.drop(columns=['Quantity', 'Month of Sourcing'])
y_train  = train_data['Quantity']







```

```python
# Step 2: Define the LSTM model
model = Sequential()
model.add(LSTM(units=100, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Step 3: Train the model
# Reshape input data for LSTM
X_train = X_train.values.reshape((X_train.shape[0], X_train.shape[1], 1))

# Fit the model
model.fit(X_train, y_train, epochs=50, batch_size=32)

```

```python
import pandas as pd
import calendar
import plotly.graph_objects as go

# Assuming df is your DataFrame and grouping_columns_unique contains the specified columns
grouping_columns_unique = ['ProductType', 'Manufacturer', 'Area Code', 'Sourcing Channel', 'Product Size', 'Product Type']

# Group by unique combinations and aggregate mean Sourcing Cost and sum Quantity
grouped = df.groupby(grouping_columns_unique + ['Year', 'Month']).agg({'Sourcing Cost':'mean', 'Quantity':'sum'}).reset_index()

# Merge with unique_combinations to retain only those combinations
merged = pd.merge(grouped, unique_combinations, on=grouping_columns_unique, how='inner')

# Create an empty figure
fig = go.Figure()

# Create an empty list to store traces for legend entries
legend_traces = []

# Iterate over unique combinations and add traces to the figure
for index, row in merged.iterrows():
    product_label = f"{row['ProductType']} - {row['Manufacturer']} - {row['Area Code']} - {row['Sourcing Channel']} - {row['Product Size']} - {row['Product Type']}"
    unique_months = row['Month']
    unique_years = row['Year']
    if isinstance(unique_months, int):  # Handle the case when unique_months is a single integer
        unique_months = [unique_months]
    monthly_data = merged[(merged['ProductType'] == row['ProductType']) & 
                          (merged['Manufacturer'] == row['Manufacturer']) & 
                          (merged['Area Code'] == row['Area Code']) & 
                          (merged['Sourcing Channel'] == row['Sourcing Channel']) & 
                          (merged['Product Size'] == row['Product Size']) & 
                          (merged['Product Type'] == row['Product Type'])]
    
    x_values = [calendar.month_name[month] for month in unique_months]
    sourcing_cost_values = monthly_data['Sourcing Cost']
    quantity_values = monthly_data['Quantity']
    
    # Calculate monthly differences
    sourcing_cost_diff = sourcing_cost_values.diff().fillna(0)
    quantity_diff = quantity_values.diff().fillna(0)
    
    # Add trace for the combination
    fig.add_trace(go.Scatter(x=x_values, y=sourcing_cost_diff, mode='lines+markers', 
                             name=f'{product_label} - Sourcing Cost', legendgroup=product_label))
    fig.add_trace(go.Scatter(x=x_values, y=quantity_diff, mode='lines+markers', 
                             name=f'{product_label} - Quantity', legendgroup=product_label))
    
    # Create an invisible trace for the legend entry
    legend_trace = go.Scatter(x=[None], y=[None], mode='markers', marker=dict(color='rgba(0,0,0,0)'),
                              name=f'{product_label}', showlegend=True, legendgroup=product_label)
    legend_traces.append(legend_trace)

# Add invisible traces to show legend entries
for trace in legend_traces:
    fig.add_trace(trace)

# Update layout
fig.update_layout(title='Change in Sourcing Cost and Quantity Over Months',
                  xaxis_title='Month',
                  yaxis_title='Change',
                  legend=dict(orientation="v", yanchor="top", y=1.0, xanchor="right", x=1),
                  )

# Show plot
fig.show()
```
